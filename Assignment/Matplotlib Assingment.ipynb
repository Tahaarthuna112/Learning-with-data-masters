```python
Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in a wide range of formats. It is widely used for data visualization and is an essential tool in the data science and scientific computing communities. Matplotlib provides a comprehensive set of tools for constructing various types of plots and charts.

Here are five types of plots that can be created using the Pyplot module of Matplotlib:

1.Line Plot: Line plots are used to display data points connected by lines. They are commonly used for visualizing trends over a continuous interval, such as time series data.

2.Scatter Plot: Scatter plots are used to display individual data points as markers on a 2D plane. They are useful for visualizing relationships and correlations between two variables.

3.Bar Plot: Bar plots are used to represent data using rectangular bars or columns, with the height or length of each bar indicating the value of a specific variable. They are often used for comparing categories or showing distribution.

4.Histogram: Histograms are used to represent the distribution of a single variable by dividing the data into bins and counting the number of data points in each bin. They are useful for understanding the data's underlying distribution.

5.Pie Chart: Pie charts are used to represent data as slices of a circular pie. Each slice represents a category or a proportion of the whole, making it easy to visualize relative proportions.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
The subplot() function in Matplotlib is used to create multiple subplots within a single figure. It allows you to arrange and display multiple plots in a grid or other custom layout. This is particularly useful when you want to compare or visualize multiple datasets or aspects of your data in a single figure.
```

```python
plt.subplot(2,2,1)
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x1,y1)
plt.title('Line 1 Plot')

plt.subplot(2,2,2)
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x2,y2)
plt.title('Line 2 Plot')

plt.subplot(2,2,3)
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x3,y3)
plt.title('Line 3 Plot')

plt.subplot(2,2,4)
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x4,y4)
plt.title('Line 4 Plot')
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
Barplot is used to show the relationship between the numeric and categoric values. In a bar chart, we have one axis representing a particular category of the columns and another axis representing the values or count of the specific category. Barcharts are plotted both vertically and horizontally and are plotted using the following line of code:

plt.bar(x,height,width,bottom,align)
x: representing the coordinates of the x-axis

height: the height of the bars

width: width of the bars. Its default value is 0.8

bottom: It’s optional. It is a y-coordinate of the bar its default value is None

align: center, edge its default value is center
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company , profit)
plt.show()
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company , profit)
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
A Box plot is used to show the summary of the whole dataset or all the numeric values in the dataset. The summary contains minimum, first quartile, median, third quartile, and maximum. Also, the median is present between the first and third quartile. Here x-axis contains the data values and y coordinates show the frequency distribution.

Parameters used in box plots are as follows:

data: NumPy array

vert: It will take boolean values i.e true or false for the vertical and horizontal plot default is True

width: This will take array and sets of the width of boxes, Optional parameters

Patch_artist: It is used to fill the boxes with color and its default value is false

labels: Array of strings which is used to set the labels of the dataset
```

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
labels = ['Box 1', 'Box 2']
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=labels, vert=False, patch_artist=True)
plt.xlabel('Value')
plt.ylabel('Box')
plt.title('Box Plot of Box 1 and Box 2')
plt.show()
```
